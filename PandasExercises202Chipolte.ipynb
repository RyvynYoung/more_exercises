### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
```

```python
# clean and change item_price to float
# exobrain https://www.kite.com/python/answers/how-to-convert-a-pandas-dataframe-column-of-strings-to-floats-in-python
chipo.item_price = chipo.item_price.str.replace('$', '')
# example: df["A"] = pd.to_numeric(df["A"], downcast="float")
#chipdf['item_price'] = pd.to_numeric(chipdf['item_price'], downcast='float')
chipo['item_price'] = pd.to_numeric(chipo['item_price'], downcast='float')
chipo.head()
```

```python
chipo['individual_price'] = chipo.item_price/chipo.quantity
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
over10 = chipo.item_name[chipo.individual_price > 10].nunique()
over10
```

```python
totalover10 = chipo.item_name[chipo.item_price > 10].nunique()
totalover10
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
# note the item_price column = individual_price * quantity
# my solution produces df with item_name and the individual price of the item
itemandprice = chipo.drop(columns=['order_id', 'quantity', 'choice_description', 'item_price'])
itemandprice
```

### Step 6. Sort by the name of the item

```python
itemandprice.sort_values(by='item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
# student_df.groupby(['Gender', 'Handed']).Ageyears.agg(['mean', 'median'])
chipo.groupby(['item_name']).individual_price.agg(['max', 'count']).sort_values(by='max', ascending=False)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggiebowl = chipo[chipo.item_name == 'Veggie Salad Bowl'].sum()
veggiebowl
```
