Day 1 - Puzzle 1

Before you leave, the Elves in accounting just need you to fix your expense report (your puzzle input); apparently, something isn't quite adding up.

Specifically, they need you to find the two entries that sum to 2020 and then multiply those two numbers together.

https://adventofcode.com/2020/day/1/input

```python
puz_input = pd.read_clipboard(names=['num'])
```

```python
puz_input.to_csv('day1_puzzle1.csv')
```

```python
puz = pd.read_csv('day1_puzzle1.csv')
```

```python
puz_list = puz.num.to_list()
puz_list
```

```python

for n in puz_list:
    this_item = n
    for i in range(0, len(puz_list)):
        next_item = puz_list[i]
        res = this_item + next_item
        if res == 2020:
            print('match!')
            print(this_item)
            print(next_item)
            ans = this_item * next_item
            print(ans)
```

Day 1 - Puzzle 2

In your expense report, what is the product of the three entries that sum to 2020?

```python
for n in puz_list:
    this_item = n
    #print(n)
    for i in range(0, len(puz_list)):
        next_item = puz_list[i]
        #print(next_item)
        for x in range(0, len(puz_list)):
            nn_item = puz_list[x]
            #print(nn_item)
            res = this_item + next_item + nn_item
            # print(res)
            if res == 2020:
                print('match!')
                print(this_item)
                print(next_item)
                print(nn_item)
                ans = this_item * next_item * nn_item
                print(ans)
```

Day 2 - Puzzle 1

Each line gives the password policy and then the password. The password policy indicates the lowest and highest number of times a given letter must appear for the password to be valid. For example, 1-3 a means that the password must contain a at least 1 time and at most 3 times.

https://adventofcode.com/2020/day/2/input

```python
d2p1 = pd.read_clipboard(names=['range', 'letter', 'password'])
```

```python
d2p1.letter = d2p1.letter.str.replace(':', '')
rangecol = d2p1.range.str.split('-', expand=True)
rangecol.columns = ['minocc', 'maxocc']
tidy_df = pd.concat([d2p1, rangecol], axis=1).drop(columns='range')

tidy_df
```

```python
tidy_df.to_csv('day2_puzzle1.csv')
```

```python
p1d2 = pd.read_csv('day2_puzzle1.csv', index_col=0)
p1d2
```

```python
letter = p1d2.letter[1]
password = p1d2.password[1]
len(re.findall(letter, password))
```

```python
len(p1d2)
```

```python
lcnt = []
for i in range(0, len(p1d2)):
    lt = p1d2.letter[i]
    pswrd = p1d2.password[i]
    cnt = len(re.findall(lt, pswrd))
    lcnt.append(cnt)
```

```python
p1d2['ltocc'] = lcnt
```

```python

p1d2
```

```python
# is lcnt > minocc and < maxocc?
p1d2['tfmin'] = np.where(p1d2.ltocc >= p1d2.minocc, 1, 0)
p1d2['tfmax'] = np.where(p1d2.ltocc <= p1d2.maxocc, 1, 0)
p1d2
```

```python
p1d2['good_pass'] = p1d2.tfmin + p1d2.tfmax
p1d2
```

```python
p1d2[p1d2.good_pass ==2].count()
```

Day 2 - Puzzle 2

Each policy actually describes two positions in the password, where 1 means the first character, 2 means the second character, and so on. (Be careful; Toboggan Corporate Policies have no concept of "index zero"!) Exactly one of these positions must contain the given letter. Other occurrences of the letter are irrelevant for the purposes of policy enforcement.

```python
p2d2 = pd.read_csv('day2_puzzle1.csv', index_col=0)
p2d2 = p2d2.rename(columns={'minocc': 'pos1', 'maxocc':'pos2'})
p2d2
```

```python
p2d2['idx1'] = p2d2.pos1 - 1
p2d2['idx2'] = p2d2.pos2 - 1
p2d2
```

```python
lt = p2d2.letter[2]
password = p2d2.password[2]
loc1 = p2d2.idx1[2]
loc2 = p2d2.idx2[2]

lt == password[loc1]
```

```python
first_true = []
for i in range(0, len(p2d2)):
    lt = p2d2.letter[i]
    pswrd = p2d2.password[i]
    loc1 = p2d2.idx1[i]
    if lt == pswrd[loc1]:
        first_true.append(1)
    else:
        first_true.append(0)
```

```python
p2d2['frst_t'] = first_true
```

```python
sec_true = []
for i in range(0, len(p2d2)):
    lt = p2d2.letter[i]
    pswrd = p2d2.password[i]
    loc2 = p2d2.idx2[i]
    if lt == pswrd[loc2]:
        sec_true.append(1)
    else:
        sec_true.append(0)
```

```python
p2d2['sec_t'] = sec_true
```

```python
p2d2
```

```python
p2d2['pos_sum'] = p2d2.frst_t + p2d2.sec_t
p2d2
```

```python
p2d2[p2d2.pos_sum ==1].count()
```
